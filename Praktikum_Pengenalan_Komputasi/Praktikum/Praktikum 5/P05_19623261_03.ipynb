```python
# Import dan inisialisasi
import pandas as pd
import matplotlib as plt
df = pd.read_csv("P05-01.csv")
```

```python
df1 = df.loc[df["Card_Category"] == "Platinum"]
df1[["Customer_Age"]].plot(kind="hist", bins=[0,20,40,60,80,100], rwidth = 0.8)
```

```python
# Nomor 2
df2 = df.loc[df["Credit_Limit"] > 5000]
df2["Income_Category"].value_counts().plot(kind = "pie")
```

```python
# Nomor 3
df1 = df.loc[df["Gender"] == "F"]
df1.plot(kind="scatter", x="Months_on_book", y="Avg_Utilization_Ratio")
```

```python
# Nomor 4
df1 = df["Attrition_Flag"].value_counts()
df1.plot(kind="barh")
```

```python
# Nomor 5
df5 = df.loc[(df["Marital_Status"] == "Married") | (df["Marital_Status"] == "Single") | (df["Marital_Status"] == "Divorced")]
df5.groupby(["Education_Level","Marital_Status"])["CLIENTNUM"].size().unstack().plot(kind="bar", stacked=True)
```
