```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("H05.csv")
```

```python
# Nomor 1: Histogram distribusi nilai matematika
data[["nilai_mat"]].plot(kind="hist", bins=[0,20,40,60,80,100], rwidth=0.8)
plt.show()
```

```python
# Nomor 2: Diagram batang banyak mahasiswa masing-masing fakultas
data["fakultas"].value_counts().plot(kind='barh')
```

```python
# Nomor 3: Diagram pie banyak mahasiswa masing-masing fakultas
data["fakultas"].value_counts().plot(kind='pie')
```

```python
# Nomor 4: Menentukan fakultas dengan banyak mahasiswa terbanyak dan menentukan diagram yang lebih baik
# Bedasarkan diagram pie dan batang horizontal, FTI adalah fakultas dengan banyak mahasiswa terbanyak
# Diagram batang horizontal menampilkan data lebih baik karena diagram batang memiliki approksimasi banyak data sedangkan
# diagram pie tidak memiliki approksimasi banyak data 
```

```python
# Nomor 5: Membuat Scatter Plot dengan x sebagai nilai kimia dan y sebagai nilai fisika
data.plot(kind='scatter', x= "nilai_kim", y='nilai_fis')
```
